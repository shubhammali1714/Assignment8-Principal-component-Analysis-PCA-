```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
# Import the data
data = pd.read_csv("/content/wine.csv")
```

```python
data1 = data.iloc[:, 1:]
```

```python
data1.head()
```

```python
data1.describe()
```

```python
data1.shape
```

```python
data1.info()
```

```python
data_array = data1.values
data_array
```

```python
data_norm = scale(data_array)
data_norm
```

```python
# implementation of PCA
pca = PCA(n_components = 13)
data_pca = pca.fit_transform(data_norm)

data_pca
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
# variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA 
plt.plot(var1,color='red')
```

```python
# Final Dataframe
final_df=pd.concat([data['Type'],pd.DataFrame(data_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCA
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
# Hierarchical Clustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
#  create Dendrograms
plt.figure(figsize=(16,12))
dendrogram=sch.dendrogram(sch.linkage(data_norm,'complete'))
```

```python
# Preparing Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(data_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Add clusters to data
data3=data.copy()
data3['clustersid']=hclusters.labels_
data3
```

 K-Mean Clustering

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot  Elbow graph for choosing K.
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm  K=3
clusters3=KMeans(3,random_state=30).fit(data_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to  data
data4=data.copy()
data4['clusters3id']=clusters3.labels_
data4
```
